```python
import pandas as pd

# Since the environment doesn't support direct file access, this code is illustrative.
# You would run this script in an environment where you have access to the Excel file.

# Load the Excel file
excel_path = 'problem-data.xlsx'  # Replace with the path to your Excel file
df_courses = pd.read_excel(excel_path, sheet_name='Courses')  # Replace with your actual sheet name
df_courses.tail(5)
```

```python
# Normalize the course names to create valid identifiers
df_courses['Course Name'] = df_courses['Course Name'].str.upper().str.replace('[^A-Z0-9 ]+', '', regex=True)
df_courses['Course Name'] = df_courses['Course Name'].str.replace(' ', '_')

# Sort by 'Course Code' to ensure sections are in order
df_courses = df_courses.sort_values(by=['Course Code'])
df_courses.tail(5)
```

```python
# Initialize a list to store the ASP facts
asp_course_facts = []

# Iterate through each row of the dataframe and generate ASP facts
for index, row in df_courses.iterrows():
    # Create a unique section number for each course code
    section = df_courses[df_courses['Course Code'] == row['Course Code']].index.get_loc(index) + 1
    asp_course_facts.append(f"course({row['Course Code']}, {row['Course Name']}, {row['Service Course']}, {row['Course Level']}, {row['Course Type']}, {row['Capacity']}, {row['Lecture Hours']}, {section}).")

# Print or write the facts to a file
print('\n'.join(asp_course_facts))
```
